Con lo aprendido en clase, de cada uno implementa un programa en Python con la solución:

- Una sala de cine cobra diferentes precios de entrada según la edad de la persona. Si una persona es menor de 3 años, la entrada es gratuita; si tienen entre 3 y 12, el boleto es de 10 pesos ; y si son mayores de 12 años, el boleto cuesta $ 15. Escribe un bucle en el que se solicite a los usuarios su edad y despliegue el costo de la entrada.

```python
edad = int(input())
if(edad < 0 ):
  print("Edad inválida")
elif(edad < 3):
  print("Entrada gratuita")
elif(edad <= 12):
  print("Costo de $10")
else:
  print("Costo de $15")
```


- Crea un diccionario que contenga la información de una canción (artista, álbum, duración y género) y despliega el contenido de cada elemento (keys, values y tuplas)

```python
yesterday = {
    "artista": "The beatles",
    "album": "Help!",
    "duracion": "2:05",
    "genero": "chamber pop"
}

print("Keys:", yesterday.keys())
print("Values: ", yesterday.values())
print("Tupla: ", yesterday.items())
```
