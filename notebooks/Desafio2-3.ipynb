Escribe un programa en Python que realice lo siguiente:

- Codifica una función que reciba una lista como argumento y que genere un diccionario como salida.

- El diccionario generado debe tener las siguientes llaves (claves): nombre, apellido y edad.


```python
import csv

def crear_dict_persona(datos_lista):
    return {
        "nombre": datos_lista[0],
        "apellido": datos_lista[1],
        "edad": datos_lista[2]
    }
```

- Genera la lista de entrada con valores ficticios de posibles personas reales.

```python
persona1 = ["Marco", "Rubio", 22]
```

```python
persona2 = ["Christian", "Granados", 22]
```

```python
persona3 = ["Pablo", "Fernandez", 13]
```

- Ejecuta el programa tres veces con valores diferentes y almacena todos los resultados en la estructura de datos más adecuada para la situación planteada.

```python
lista_diccionarios = []
lista_diccionarios.append(crear_dict_persona(persona1))
lista_diccionarios.append(crear_dict_persona(persona2))
lista_diccionarios.append(crear_dict_persona(persona3))
print(lista_diccionarios)
```

- Escribe el diccionario del ejercicio anterior en un archivo CSV o JSON.

```python
nombre_archivo = "personas.csv"
campos = ["nombre", "apellido", "edad"]
```

```python
with open(nombre_archivo, mode="w", newline="", encoding="utf-8") as archivo_csv:
    escritor = csv.DictWriter(archivo_csv, fieldnames=campos)

    escritor.writeheader()  # Escribe la fila de títulos
    escritor.writerows(lista_diccionarios)

print(f"Archivo '{nombre_archivo}' generado con éxito.\n")
```

- Usando alguna de las funciones de lectura de Python, despliega el contenido de lo anterior.

```python
print("--- Contenido leído del archivo CSV ---")
with open(nombre_archivo, mode="r", encoding="utf-8") as archivo_lectura:
    lector_dict = csv.DictReader(archivo_lectura)
    for fila in lector_dict:
        print(f"Persona: {fila['nombre']} {fila['apellido']}, Edad: {fila['edad']}")
```

- Por medio de la clase persona, modela la información del ejercicio anterior y crea una instancia que contenga tu nombre, apellido y edad como atributos.

```python
class Persona:
    def __init__(self, nombre, apellido, edad):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad

    def __str__(self):
        return f"Instancia de Persona -> {self.nombre} {self.apellido}, {self.edad} años."

# Mi instancia personalizada
mi_instancia = Persona("Guillermo", "Valencia", 21)
print("\n--- Modelado de Clase ---")
print(mi_instancia)
```
